```python
import tensorflow as tf
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../dataset/pokemon.csv', index_col='Number')
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns.to_list()
```

```python
df = df[['isLegendary','Generation', 'Type_1', 'Type_2', 'HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed','Color','Egg_Group_1','Height_m','Weight_kg','Body_Style']]
```

```python
df
```

```python
df['isLegendary'] = df['isLegendary'].astype(int)
```

```python
df
```

```python
df.describe(include='all')
```

```python
pd.get_dummies(df['Color']) # we will use this
```

```python
def dummy_creator(df,cols):
    for col in cols:
        df_dummy = pd.get_dummies(df[col])
        df = pd.concat([df,df_dummy],axis=1)
        df = df.drop(col,axis=1)
    return df
```

```python
cleandf =  dummy_creator(df,['Color','Egg_Group_1','Body_Style','Type_1','Type_2'])
```

```python
cleandf
```

```python
X = cleandf.iloc[:,1:]
y = cleandf.iloc[:,0]
print(X.shape)
print(y.shape)
```

```python
scaler = preprocessing.StandardScaler()
x = scaler.fit_transform(X)
```

```python
trainx, testx, trainy, testy = train_test_split(x,y)
```

## simple ANN model

```python
inputshape =   trainx.shape[1]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(500,activation='relu',input_shape=(inputshape,)))
model.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```python
model.compile(optimizer='sgd',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x=trainx,y=trainy,epochs=28,validation_data=(testx,testy))
```

```python
loss_val,accuracy_val = model.evaluate(testx,testy)
print(loss_val)
print(accuracy_val)
```

```python
plt.figure(figsize=(20,5))
plt.plot(history.history['accuracy'],label='train accuracy')
plt.plot(history.history['val_accuracy'],label='test accuracy')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(history.history['loss'],label='train loss')
plt.plot(history.history['val_loss'],label='test loss')
plt.show()
```

```python
model.save('../models/pokemon_islegendary_predictor_model.h5')
```
